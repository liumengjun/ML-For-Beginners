```python
# Load the hotel reviews from CSV
import pandas as pd
import time
# importing time so the start and end time can be used to calculate file loading time
print("Loading data file now, this could take a while depending on file size")
start = time.time()
# df is 'DataFrame' - make sure you downloaded the file to the data folder
df = pd.read_csv('../data/Hotel_Reviews.csv')
end = time.time()
print("Loading took " + str(round(end - start, 2)) + " seconds")
```

```python
top_10_nation_df = df.where(df['Reviewer_Nationality'].isin(top_10_nationalities.index))
top_10_nation_df
```

```python
top_10_nation_df['Hotel_Name'].value_counts().head(20)
```

```python
top_10_nation_hotels_df = (top_10_nation_df[['Reviewer_Nationality','Hotel_Name']]).value_counts().reset_index()
top_10_nation_hotels_df
```

```python
# top_10_nation_hotels_df.columns = ['Reviewer_Nationality', 'Hotel_Name', 'count']
top_10_nation_hotels_df
```

```python
top_10_nation_hotels_df.sort_values(['Reviewer_Nationality', 'count'], ascending=[True, False], inplace=True)
top_10_nation_hotels_df['Reviewer_Nationality2'] = top_10_nation_hotels_df['Reviewer_Nationality'].shift()
top_10_nation_hotels_df
```

```python
top_10_nation_hotels_df[top_10_nation_hotels_df['Reviewer_Nationality2'] !=  top_10_nation_hotels_df['Reviewer_Nationality']]
```

```python
top_10_nation_hotels_df['row_number'] = top_10_nation_hotels_df.groupby('Reviewer_Nationality')['count'].rank(ascending=False,method='dense')
top_10_nation_hotels_df[top_10_nation_hotels_df['row_number']==1]
```

```python
hotel_freq_df = df.loc[:,['Hotel_Name','Total_Number_of_Reviews']]
hotel_freq_df['Total_Reviews_Found'] = hotel_freq_df.groupby('Hotel_Name').transform('count')
hotel_freq_df = hotel_freq_df.drop_duplicates(subset = ["Hotel_Name"])
hotel_freq_df
```

```python
review_scores_df = df[['Hotel_Name','Average_Score','Reviewer_Score']].groupby('Hotel_Name').mean()
review_scores_df['Calc_Average_Score'] = review_scores_df["Reviewer_Score"].round(1)
review_scores_df['Average_Score_Difference'] = review_scores_df["Average_Score"] - review_scores_df["Calc_Average_Score"]
review_scores_df = review_scores_df.sort_values(by=["Average_Score_Difference"])
review_scores_df
```
